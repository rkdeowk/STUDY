```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.read_csv('./wafer_23012020_041211.csv', index_col=0)
```

```python
# Thickness 만들기
def map_value(x):
    if x == 1:
        return random.uniform(118, 122)
    else:
        random_number = random.uniform(117, 124) 
        while 118 <= random_number <= 122: 
            random_number = random.uniform(117, 124)
        return random_number
    
df['Thickness'] = df['Good/Bad'].apply(map_value)
df.head()
```

```python
df.isna().sum()
```

```python
# drop na
df_dropna = df.dropna(axis=1)
df_dropna.head()
```

```python
# unique 값이 1개인 것 check
# 모두 같은 값이면 독립
unique_is_one_list = []
for col in df_dropna.columns:
    if df_dropna[col].nunique() == 1:
        unique_is_one_list.append(col)

len(unique_is_one_list)
```

```python
for col in unique_is_one_list:
    print(df_dropna[col].unique())
```

```python
unique_value_counts = df_dropna.nunique()
to_remove = unique_value_counts[unique_value_counts == 1].index
df_dropna = df_dropna.drop(columns=to_remove)
df_dropna
```
