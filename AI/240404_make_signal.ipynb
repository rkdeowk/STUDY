```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for i in range(1, 6):
    df = pd.read_csv(f'./datasets/trend/trend{i}.csv')
    plt.figure(figsize=(20, 3))
    plt.title(i)
    plt.plot(df['MAX'])
    plt.tight_layout()
    plt.show()
```

```python
def downsample_signal(signal, sampling_interval):
    sample_indices = np.arange(0, len(signal), sampling_interval)
    downsampled_signal = [max(signal[i:i+sampling_interval]) for i in sample_indices]
    return downsampled_signal
```

```python
for i in range(1, 6):
    df = pd.read_csv(f'./datasets/trend/trend{i}.csv')
    data = downsample_signal(df['MAX'], 100)
    plt.figure(figsize=(20, 3))
    plt.title(i)
    plt.plot(data)
    plt.tight_layout()
    plt.show()
```

```python
window_size = 16
for i in range(1, 6):
    df = pd.read_csv(f'./datasets/trend/trend{i}.csv')
    signal_original = df['MAX']
    signal_original = pd.Series(downsample_signal(signal_original, 10))
    signal_moving_average = pd.concat([pd.Series([np.nan] * (window_size - 1)), signal_original])
    signal_moving_average = signal_moving_average.rolling(window=window_size).mean()
    print(len(signal_original), len(signal_moving_average))
    plt.figure(figsize=(10, 3))
    plt.plot(signal_original, label='Original Signal', marker='o')
    plt.plot(signal_moving_average, label=f'Moving Average (Window Size = {window_size})', color='red')
    plt.title(f'{i}')
    plt.legend()
    plt.show()
```

```python
i = 1
window_size = 16
downsample_size = 20
df = pd.read_csv(f'./datasets/trend/trend{i}.csv')
signal_original = df['MAX']
signal_original = pd.Series(downsample_signal(signal_original, downsample_size))
signal_moving_average = pd.concat([pd.Series([np.nan] * (window_size - 1)), signal_original])
signal_moving_average = signal_moving_average.rolling(window=window_size).mean()
print(len(signal_original), len(signal_moving_average))
plt.figure(figsize=(10, 3))
plt.plot(signal_original, label='Original Signal', marker='o')
plt.plot(signal_moving_average, label=f'Moving Average (Window Size = {window_size})', color='red')
plt.title(f'{i}')
plt.legend()
plt.show()
```
